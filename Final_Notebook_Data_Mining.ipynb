```python
x = data.drop('Class', axis = 'columns')
y = data['Class']
```

```python
smote = SMOTE(sampling_strategy = 'minority')
x_sm, y_sm = smote.fit_resample(x,y)
```

```python
No_cases= len(data)
non_fraud_cases = len(data[data.Class==0])
Fraud_cases= len(data[data.Class==1])
Fraud_percentage=round(Fraud_cases/non_fraud_cases*100,2)

print(cl('Dataset count',attrs=['bold']))
print(cl('------------------------------------'))
print(cl('Total number of cases are {}'.format(No_cases),attrs=['bold']))
print(cl('Total number of non-fraud cases are {}'.format(non_fraud_cases), attrs=['bold']))
print(cl('Total number of fraud cases are {}'.format(Fraud_cases), attrs=['bold']))
print(cl('Percentage of fraud cases is {}'.format(Fraud_percentage), attrs=['bold']))
```

```python
model = ExtraTreesClassifier()
model.fit(x_sm, y_sm)
model.feature_importances_
feat = pd.Series(model.feature_importances_, index = x_sm.columns)
```

```python
plt.figure(figsize = (15,15))
feat.nlargest(50).plot(kind = 'barh',color='Red')
```

```python
plot = feat.nlargest(18)
plot.index
```

```python
cols = ['V14', 'V11', 'V12', 'V17', 'V4', 'V10', 'V3', 'V16', 'V18', 'V9', 'V1',
       'V2', 'V7', 'V19', 'Time', 'V8', 'V21', 'V13']
```

```python
x_new = x_sm[cols]
```

```python
x_sm.shape
```

```python
x_new.shape
```

```python
#spliting the dataset into 70% and 30% model
X_train, X_test, y_train, y_test = train_test_split(x_new, y_sm, train_size=0.7,test_size = 0.3, random_state = 0)
```

```python
randomf = RandomForestClassifier()
randomf.fit(X_train, y_train)
random_pred = randomf.predict(X_test)
random_pred.shape
```

```python
print("Random forest tree accuracy is {}:" .format(accuracy_score(y_test, random_pred)))
```

```python
Decision_model = DecisionTreeClassifier(max_depth = 2, criterion = 'entropy')
Decision_model.fit(X_train, y_train)
Decision_pred = Decision_model.predict(X_test)
Decision_pred.shape
```

```python
print("Decision tree accuracy is {}:" .format(accuracy_score(y_test, Decision_pred)))
```

```python
NB= GaussianNB()
NB.fit(X_train, y_train)
NB_pred= NB.predict(X_test)
NB_pred.shape
```

```python
#Logistic Regression
LR= LogisticRegression()
LR.fit(X_train,y_train)
LR_pred=LR.predict(X_test)
```

```python
print("Logistic Regression model accuracy is {}:" .format(accuracy_score(y_test, LR_pred)))
```

```python
table=[['Algorithm','Accuracy'],['Decision Tree',(accuracy_score(y_test, Decision_pred))],['Random Forest Tree',accuracy_score(y_test, random_pred)],['Logistic Regression',accuracy_score(y_test, LR_pred)],['Naive Bayes',accuracy_score(y_test, NB_pred)]]
```

```python
print(tabulate(table))
```
